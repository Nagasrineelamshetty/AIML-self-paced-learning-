```python
from types import MemberDescriptorType
import pandas as pd
import numpy as np
np.random.seed(1212)
import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers

```

```python
df_train= pd.read_csv('mnist_train.csv')
df_test=pd.read_csv('mnist_test.csv')
```

```python
df_train.head()#28*28=784
```

```python
df_test.head()
```

```python
df_features= df_train.iloc[:,1:785]
df_label= df_train.iloc[:,0]
X_test= df_test.iloc[:,0:784]
print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_cv,y_train,y_cv= train_test_split(df_features,df_label,test_size=0.2,random_state=1212)
X_train= X_train.to_numpy().reshape(48000,784)
X_cv= X_cv.to_numpy().reshape(12000,784)#Validation
X_test=X_test.to_numpy().reshape(10000,784)
```

```python
df_features.shape[0]
```

```python
print((min(X_train[1]),max(X_train[1])))
```

```python
#Feature Normalization
X_train= X_train.astype('float32');X_cv= X_cv.astype('float32');X_test=X_test.astype('float32')
X_train/=255;X_cv/=255;X_test/=255
#Convert labels into one hot encoding
num_digits=10
y_train=keras.utils.to_categorical(y_train,num_digits)
y_cv=keras.utils.to_categorical(y_cv,num_digits)
```

```python
#Printing two examples of labels after conversion
print(y_train[0])
print(y_train[3])
```

```python
n_input= 784
n_hidden_1=300
n_hidden_2=100
n_hidden_3=100
n_hideen_4=200
num_digits=10
```

```python
Inp= Input(shape=(784,))
x= Dense(n_hidden_1,activation='relu',name='Hidden_Layer_1')(Inp)
x= Dense(n_hidden_2,activation='relu',name='Hidden_Layer_2')(x)#Connecting layers
x= Dense(n_hidden_3,activation='relu',name='Hidden_Layer_3')(x)
x= Dense(n_hideen_4,activation='relu',name='Hidden_Layer_4')(x)
output= Dense(num_digits,activation='softmax',name='Output_Layer')(x)#softmax gives probabilities which amount to 1
```

```python
#Our model should have 6 layers - input+4+output layer
model= Model(Inp,output)
model.summary()
```

```python
#Insert hyperparameters
learning_rate=0.1
training_epochs=20
batch_size=100
sgd=optimizers.SGD(learning_rate=learning_rate)

```

```python
#We use Plain Vanilla Stochastic Gradient Descent as our optimizing methodology
model.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history1=model.fit(X_train,y_train,batch_size=batch_size,epochs=training_epochs,verbose=2,validation_data=(X_cv,y_cv))
```
