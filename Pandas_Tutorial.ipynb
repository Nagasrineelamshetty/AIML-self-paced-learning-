###Data Frames

```python
import numpy as np
```

```python
data= {
    "Name":["Nagasri","Charan","Malathi","Srinivas"],
    "Age": [19,15,53,54],
    "Department": ["Education","Business","Teaching","Business"],
    "Salary": [500000,60000,np.nan,90000]
}
```

```python
data= pd.DataFrame(data)
print(data)
```

```python
data.head()
```

```python
data.tail()
```

```python
#loc and iloc
print(data.iloc[1:3])
```

```python
print(data.loc[1:2,["Age","Department"]])
print(data.iloc[1:3, 1:3])#row,cloumn
```

```python
data[["Age","Department"]]
```

```python
data.drop("Age",axis=1)#inplace= true deletes from the dataframe itself
```

```python
print(data)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
#Broadcasting
data["Salary"]= data["Salary"]+ 5000#Broadcasted to match the shape and size of a particular column
```

```python
data["Salary"]
```

```python
#Rename columns
data.rename(columns={"Department":"Dept"},inplace= True)
print(data)
```

```python
data["Salary"].unique()
```

```python
data["Dept"].unique()
```

```python
data["Dept"].value_counts()
```

```python
data["Promoted Salary"]= data["Salary"]*10
print(data)
```

```python
#Data Cleaning
data.isnull().sum()
```

```python
data.dropna(how="any")# any row that had any null value
```

```python
data.dropna(how="all")# all rows that have null values which we don't have
```

```python
#Other way is to fill the values
data["Age"].fillna(data["Age"].mean())
```

```python
data["Salary"].fillna(data["Salary"].median())
```

```python
#forward fill and backward fill
data["Age"].fillna(method='ffill')
```

```python
data["Age"].fillna(method='bfill')
```

```python
data["Name"].replace("Nagasri","Nagasri_Neelamshetty",inplace=True)
data["Name"].replace("Malathi","Malathi_Neelamshetty")
data["Name"].replace("Charan","Charan_Neelamshetty")
```

```python
print(data)
```

```python
#Dealing with duplicate values
data_dup= data[data.duplicated(keep="first")]#keep= first first element is kept if keep= last last element is kept
print(data_dup)
```

```python
#Invalid Values:
#Lambda
data["Promoted Salary"]= data["Promoted Salary"].apply(lambda x: x/10 if x> 650000 else x)
```

```python
#name= "alice_fernandes"
data[["first_name","last_name"]]= data["Name"].str.split("_",expand= True,n=1)#splits data based on '_'
print(data)
```

```python
#apply and lambda
def multiplying_age(x):
  return x*2
data["Age"]= data["Age"].apply(multiplying_age)
print(data)
```

```python
data["Age"]= data["Age"].apply(lambda x: x/2)
print(data)
```

```python
data["Age"]= data["Age"].apply(lambda x: x*2)
print(data)
```

```python
#Joins and Merges
#left,right, outer and inner
#merge- common one once and all other properties
department_info= {
    "Dept": ["HR","IT","Finance"],
    "Location": ["NewYork","SanFrancisco","Chicago"],
    "Manager": ["Laura","Steve","Nina"]
}
df2= pd.DataFrame(department_info)
print(df2)
```

```python
pd.concat([data,df2])#Stacked Horizontally
```

```python
pd.concat([data,df2], axis=1) #Stacked Vertically
```

```python
pd.merge(data, df2, on="Dept")
```
